```python
import subprocess, time

process = subprocess.Popen(
    ['python', 'server.py', '--model', 'qwen2.5-coder-7b', '--loader', 'llama.cpp', '--n-gpu-layers', '100', '--api', '--listen'],
    stdout=subprocess.PIPE, stderr=subprocess.PIPE
)
time.sleep(60)  # sačekaj da se 7B model učita (videli smo da traje ~55 sekundi)
print("Server pokrenut u pozadini")
```

```python
print(process.poll())  # None = radi; broj = ugasio se
```

```python
import requests, time

url = "http://127.0.0.1:5000/v1/completions"
payload = {
    "prompt": "Napisi Python funkciju koja proverava da li je broj prost, sa objasnjenjem koda.",
    "max_tokens": 1000,
    "temperature": 0.8
}

start = time.time()
response = requests.post(url, json=payload)
elapsed = time.time() - start
data = response.json()

print(data["choices"][0]["text"])

# Metrike - isti princip kao Ollama verbose output
usage = data.get("usage", {})
prompt_tokens = usage.get("prompt_tokens", "N/A")
completion_tokens = usage.get("completion_tokens", "N/A")
total_tokens = usage.get("total_tokens", "N/A")

print(f"\n--- METRIKE ---")
print(f"total duration:       {elapsed:.2f}s")
print(f"prompt tokens:        {prompt_tokens}")
print(f"completion tokens:    {completion_tokens}")
print(f"total tokens:         {total_tokens}")
if isinstance(completion_tokens, int) and elapsed > 0:
    print(f"eval rate:            {completion_tokens/elapsed:.2f} tokens/s")
```

```python
process.terminate()
import time
time.sleep(5)
print("Server ugašen")
```
